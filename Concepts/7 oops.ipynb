4 pillars of OOPS
        1. Encapsulation

                Binding data + methods together

                👉 Example:

                class Student:
                def __init__(self):
                        self.name = "Manthan"

                ✔ Data is wrapped inside class

        2. Abstraction

                Hiding unnecessary details, showing only important parts

                👉 Example:

                You use a car → don’t know engine details

                In Python:

                using methods without exposing logic
        3. Inheritance

                One class uses properties of another

                👉 Example:

                class Parent:
                def show(self):
                        print("Parent")

                class Child(Parent):
                pass

                ✔ Child gets Parent features

        4. Polymorphism

                Same function behaves differently

                👉 Example:

                def add(a, b):
                return a + b

                print(add(2, 3))        # 5
                print(add("Hi ", "Bro"))  # Hi Bro

                ✔ Same function → different behavior

Inheritance
🔹 What is Inheritance?

Inheritance allows one class to use properties (variables + methods) of another class.

👉 What this really means is:
You don’t rewrite code again and again, you reuse it.

    types of Inheritance ---> 5
    1.simple - One parent → one child
    2.multiple - One child → multiple parents
    3.multilevel - Chain system (grandparent → parent → child)
    4.heirarical
    5.Hybrid - which have 2 or more than 2 Inheritance

Super Function - same constructor child class mein use krna hai jo parent mein hai 
                super.Function_name() we have to write to call a function from parent to child
                Important Notes about super()
                Used to call parent class methods
                Helps avoid rewriting code
                Works with constructors and normal functions
                Syntax:
                super().method_name()

Method Overriding :
    Child class modifies parent method


Method Resolution Order (MRO) :
    Python decides which method to run first
    jo parent pehle daala hoga voh uska chalega
    Important in multiple inheritance

```python
class Parent:
    def show(self):
        print("Parent")

class Child(Parent):
    pass
```

```python
class A:
    def a(self):
        print("Class A")

class B:
    def b(self):
        print("Class B")

class C(A, B):
    pass
```

```python

class Universe:
    def universe_method(self):
        print("universe is calling")
    
class Earth(Universe):
    def earth_method(self):
        print("earth is calling")
class India(Earth):
    def india_method(self):
        print("india is calling")
    
n=India()
n.universe_method()
n.earth_method()
n.india_method()
```

```python
#constructor Case
class parent:
    def __init__(self,msg):
        self.Msg=msg
        
    def show(self):
        print(self.Msg)

class child(parent):
    def __init__(self, msg):
        super().__init__(msg)
        # super().show()  

c=child("ctyu")
```

```python
#function Case
class parent:
    def __init__(self, name, roll):
        self.name = name
        self.roll = roll

    def show(self):
        print(self.name, self.roll)


class child(parent):
    def __init__(self, name, roll, msg):
        self.msg = msg
        super().__init__(name, roll)   # correct constructor call

    def display(self):
        super().show()   # function call from parent
        print(self.msg)


c = child("ctyu", 778, "drtfgyu")
c.display()
```

```python
#method over riding
class Parent:
    def show(self):
        print("Parent")

class Child(Parent):
    def show(self):
        print("Child")

#child case will run
```
