## 1. Reading a File

```python
# Reading a file using open()
# 'r' mode = read mode
# always close the file after reading using f.close()

f = open("harry.txt", "r")

content = f.read()    # reads entire file content as a string

print(content)

f.close()             # always close the file after use
```

## 2. Writing to a File

```python
# Write to a file called John_Doe.txt
# It should contain data about John Doe
# 'w' mode = write mode
# Note: 'w' mode creates file if not exists, overwrites if exists

f = open("John_Doe.txt", "w")

string = '''
John Doe is a nice guy. He lives in Nyc and he works with Python
His favorite package is Pandas
'''

f.write(string)

f.close()

print("File written successfully!")
```

## 3. Appending to a File

```python
# Append to an existing file called John_Doe.txt
# It should add data about John Doe's Hometown
# 'a' mode = append mode
# Note: 'a' mode adds content at the END without deleting existing content

f = open("John_Doe.txt", "a")

string = '''
John Doe initially lived in Phoenix, Arizona. He is a very nice guy
'''

f.write(string)

f.close()

print("Content appended successfully!")
```

## 4. Reading File Line by Line

```python
# Reading a file line by line using for loop
# useful when file is very large and we dont want to load all at once

f = open("harry.txt", "r")

for line in f:          # iterates one line at a time
    print(line)

f.close()
```

## 5. Using `with` Statement (Context Manager)

```python
# Old way (manual close)
# f = open("harry.txt", "r")
# content = f.read()
# print(content)
# f.close()

# Better way using 'with' statement (context manager)
# No need to write f.close() because file is automatically
# closed when the with block ends

with open("harry.txt", "r") as f:    # context manager
    content = f.read()
    print(content)
```

## 8. Argparse — Command Line Arguments

```python
# argparse is used to pass arguments from command line to python script
# Example usage (run in terminal, not notebook):
# python script.py 10 5 add  --> The result is 15.0
# python script.py 10 5 mul  --> The result is 50.0

import argparse

parser = argparse.ArgumentParser(description="Simple Calculator")

parser.add_argument("num1", type=float, help="First number")
parser.add_argument("num2", type=float, help="Second number")
parser.add_argument("operation", choices=["add", "sub", "div", "mul"], help="Operation to perform")

# Note: parse_args() works in terminal/command line
# In notebook, use parse_known_args() to avoid errors
args, _ = parser.parse_known_args()

# print(args)

if args.operation == "add":
    print(f"The result is {args.num1 + args.num2}")

elif args.operation == "sub":
    print(f"The result is {args.num1 - args.num2}")

elif args.operation == "mul":
    print(f"The result is {args.num1 * args.num2}")

elif args.operation == "div":
    print(f"The result is {args.num1 / args.num2}")

else:
    print("Some error occurred")
```

## Quick Reference

| Mode | Symbol | Purpose |
|------|--------|---------|
| Read | `'r'` | Read existing file |
| Write | `'w'` | Create/overwrite file |
| Append | `'a'` | Add to end of file |
| Read+Write | `'r+'` | Read and write |

| Module | Use |
|--------|-----|
| `os` | File/folder existence, paths, delete |
| `shutil` | Copy, move, delete folders |
| `argparse` | Command line arguments |
