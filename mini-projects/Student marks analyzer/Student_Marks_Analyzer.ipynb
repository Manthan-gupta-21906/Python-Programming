# Student Marks Analyzer

Problem Statement Develop a program that analyzes student marks using Python and
NumPy

Logic Explanation :-

Step 1 — Collecting data
First the program asks how many students there are. Then it runs a loop that many times. Inside that loop, it asks for the student's name and how many subjects they have. Then another loop runs inside to collect marks for each subject one by one.

Step 2 — Storing in NumPy array
The plain Python list is converted into a NumPy 2D array using np.array(). This organizes data into rows and columns like a table.

Step 3 — Calculations using axis=1
Four NumPy functions are applied on the array. All use axis=1 which means "operate across each row" — giving one result per student instead of per subject.

Step 4 — Grade assignment logic
A function is called for each student. It receives that student's average and checks it against conditions using if/elif/else

Step 5 — Displaying results
A loop runs through all students using index i. For each student it picks names[i], totals[i], averages[i], etc. and calls grade(averages[i]) to get the grade. Everything is printed in a formatted report.



```python
import numpy as np
```

```python
def get_student_details():
  n=int(input("enter number of students to add : "))
  names=[]
  marks=[]
  for i in range (n):
    name_=input(f"enter name of student {i+1} : ")
    subjects=int(input(f"enter number of subjects of {name_} : "))
    marks_=[]
    for j in range (subjects):
      m = float(input(f"  Marks for subject {j+1}: "))
      marks_.append(m)
    names.append(name_)
    marks.append(marks_)
  return names,marks
```

```python
def store_numpy(marks):
  marks2= np.array(marks)
  return marks2
```

```python
def calculate(marks2):
  total=np.sum(marks2,axis=1)
  average=np.mean(marks2, axis=1)
  max_=np.max(marks2, axis=1)
  min_=np.min(marks2, axis=1)
  return total,average,max_,min_
```

```python
def grade(average):
  if average >= 85:
        return "A"
  elif average >= 70:
        return "B"
  elif average >= 50:
        return "C"
  else:
        return "Fail"
```

```python
def display(names, totals, averages, highest, lowest):
    print('STUDENT MARKS REPORT : ')
    for i in range(len(names)):
        print(f"Student  : {names[i]}")
        print(f"Total    : {totals[i]:}")
        print(f"Average  : {averages[i]:}")
        print(f"Highest  : {highest[i]:}")
        print(f"Lowest   : {lowest[i]:}")
        print(f"Grade    : {grade(averages[i])}")
        print("-"*55)
```

```python
names,marks = get_student_details()
marks_ =store_numpy(marks)
total,average,max_,min_=calculate(marks_)
display(names,total,average,max_,min_)
```
